```python
import pandas as pd

melb_data = pd.read_csv(data_path + '/CSVs/melb_data_ps.csv', sep=',')
melb_data.head()
```

```python
melb_df = melb_data.copy()
melb_df = melb_df.drop(['index', 'Coordinates'], axis=1, inplace=False) #cols removement


total_rooms = melb_df['Bathroom'] + melb_df['Bedroom'] + melb_df['Rooms']

melb_df['MeanRoomsArea'] = melb_df['BuildingArea']/total_rooms


diff_area = melb_df['BuildingArea'] - melb_df['Landsize']
sum_area = melb_df['BuildingArea'] + melb_df['Landsize']
melb_df['AreaRatio'] = diff_area/sum_area
melb_df.head()
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)
melb_df['Date']

year_sold = melb_df['Date'].dt.year

print('Min year sold:', year_sold.min())
print('Max year sold:', year_sold.max())
print('Mode year sold:', year_sold.mode()[0])


melb_df['MonthSale'] = melb_df['Date'].dt.month
melb_df['MonthSale'].value_counts(normalize=True)*100


delta_days = melb_df['Date'] - pd.to_datetime('2016-01-01')
# display(delta_days.dt.days)

melb_df['AgeBuilding'] = melb_df['Date'].dt.year - melb_df['YearBuilt']
melb_df.drop(['YearBuilt'], axis=1, inplace=True)
melb_df.head()
```

```python
display(melb_df['Address'].nunique())


def get_street_type(address):
    exclude_lst = ['N', 'S', 'E', 'W']
    address_lst = address.split(' ')
    
    if address_lst[-1] in exclude_lst:
        return address_lst[-2]
    else:
        return address_lst[-1]
        
melb_df['Address'] = melb_df['Address'].apply(get_street_type)


def change_adress(string):
    alt_dict = {'Av_alt' : ['Avenue', 'Ave'],
                'Cr_alt' : ['Crescent'],
                'Bvd_alt' : ['Boulevard'],
                'Hwy_alt' : ['Highway'], 
                'Pde_alt' : ['Parade'] }
    
    for i in alt_dict:               
        if string in alt_dict[i]:
            key_lst = i.split('_')
            string = key_lst[0]
       
    return string  


melb_df['Address'] = melb_df['Address'].apply(change_adress)



melb_df['Address'].value_counts()
melb_df['Address'].nunique()
```

```python
popular_stypes = melb_df['Address'].value_counts().nlargest(10).index
display(popular_stypes)

melb_df['Address'] = melb_df['Address'].apply(lambda x: x if x in popular_stypes else 'other')
display(melb_df['Address'].nunique())
melb_df['Address'].value_counts()
```

```python
unique_list = []
# пробегаемся по именам столбцов в таблице
for col in melb_df.columns:
    # создаём кортеж (имя столбца, число уникальных значений)
    item = (col, melb_df[col].nunique(), melb_df[col].dtypes) 
    # добавляем кортеж в список
    unique_list.append(item) 

# создаём вспомогательную таблицу и сортируем её по переменной Num_Unique
unique_counts = pd.DataFrame(
    unique_list,
    columns=['Column_Name', 'Num_Unique', 'Type']
    ).sort_values(by='Num_Unique',  ignore_index=True)


#display(unique_counts)


cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'] # список столбцов, которые мы не берём во внимание
max_unique_count = 150 # задаём максимальное число уникальных категорий
for col in melb_df.columns: # цикл по именам столбцов
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude: # проверяем условие
        melb_df[col] = melb_df[col].astype('category') # преобразуем тип столбца
#display(melb_df.info())

#print(melb_df['Regionname'].cat.categories)
#print(melb_df.columns)

melb_df['Type'] = melb_df['Type'].cat.rename_categories({
    'h' : 'house',
    't' : 'townhouse',
    'u' : 'unit' 
})

melb_df['Type'].cat.categories
```

```python
melb_df['Type'] = melb_df['Type'].cat.add_categories('flat') # добавили в переменную Type доп. категорию Flat

new_houses_types = pd.Series(['unit', 'house', 'flat', 'flat', 'house']) 
new_houses_types = new_houses_types.astype(melb_df['Type'].dtype)

display(new_houses_types)
```

```python
arg_1 = 'Опыт работы 8 лет 3 месяца'
arg_2 = 'Опыт работы 3 года 5 месяцев'
arg_3 = 'Опыт работы 1 год 9 месяцев'
arg_4 = 'Опыт работы 3 месяца'
arg_5 = 'Опыт работы 6 лет'

def get_experience(arg):
    count = 0
    arg = arg.split(' ')
    
    if arg[-1].startswith(('л', 'г')):
        count += int(arg[-2])*12   
        
    elif arg[-3].startswith(('л', 'г')):
        count += int(arg[-4])*12 + int(arg[-2])
        
    elif arg[-1].startswith(('мес')):
        count += int(arg[-2])
        
    return count

    
display(get_experience(arg_1))
display(get_experience(arg_2))
display(get_experience(arg_3))
display(get_experience(arg_4))
display(get_experience(arg_5))
```

```python
# Задание 7.12 
# Создайте признак времени суток поездки time_of_day. Время суток будем определять из часа начала поездки. Условимся, что:

#    поездка совершается ночью (night), если её час приходится на интервал от 0 (включительно) до 6 (включительно) часов;
#    поездка совершается утром (morning), если её час приходится на интервал от 6 (не включительно) до 12 (включительно) часов;
#    поездка совершается днём (day), если её час приходится на интервал от 12 (не включительно) до 18 (включительно) часов;
#    поездка совершается вечером (evening), если её час приходится на интервал от 18 (не включительно) до 23 часов (включительно).

# Во сколько раз количество поездок, совершённых днём, больше, чем количество поездок, совёршенных ночью, за представленный в данных период времени? Ответ округлите до целых.

def time_of_day(time):
    if time.hour in range(0, 7): return 'night'
    elif time.hour in range(7, 13): return 'morning'
    elif time.hour in range(13, 19): return 'day'
    else: return 'evening'
        
citybike_mod['time_of_day'] = citybike_mod['stoptime'].apply(time_of_day)


day_mask = citybike_mod['time_of_day'] == 'day'
night_mask = citybike_mod['time_of_day'] == 'night'

len(citybike_mod[day_mask]) / len(citybike_mod[night_mask])
```
